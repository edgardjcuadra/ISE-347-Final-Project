```python
import re
import pandas as pd

# Replace this with the path to your input file
input_file = './ampl/out75.txt'
output_file = './data/index75.csv'

# Read the entire file
with open(input_file, 'r') as f:
    content = f.read()

# Regular expression to find objFunc and the next 99 x values
pattern = re.compile(r"objFunc:(\d+\.\d+)\s+x values:\s+((?:\d+\.\d+\s+){99})", re.MULTILINE)
matches = pattern.findall(content)

# Convert matches into structured data
data = []
for obj_func, x_vals in matches:
    x_list = list(map(float, x_vals.strip().split()))
    x_list.append(float(obj_func))  # Add objFunc at the end
    data.append(x_list)

# Create column names: x1 to x99 + objFunc
columns = [f"x{i+1}" for i in range(99)] + ['objFunc']

# Create DataFrame and save to CSV
df = pd.DataFrame(data, columns=columns)
df.to_csv(output_file, index=False)

print(f"Data extracted and saved to {output_file}")
df 
```

```python
import scipy.stats as stats

df = pd.read_excel('./data/ex_ret.xlsx')

# Normality Test (Shapiro-Wilk test)
headers = ["r10", "r25", "r50", "r75"]
sts_data = {
    'returns_q': headers, 
    'W-stat': [], 
    'W-p-stat': [], 
    't-stat': [], 
    't-p-val': [], 
}
ttest = []
wiltest = []
print("Normality Test Results: p-value > 0.05 then normal")
for column in headers:
    stat, p_value = stats.shapiro(df[column])
    if p_value > 0.05: 
        ttest.append(column) 
    else: 
        wiltest.append(column)
    print(f"{column}: W-stat = {stat:.4f}, p-value = {p_value:.4f}, {p_value>0.05}")
    sts_data['W-stat'].append(stat)
    sts_data['W-p-stat'].append(p_value)

print("\nPaired T-Test Results (vs. benchmark): p_val > 0.05 then same distribution")
for column in ttest:
    t_stat, t_p_value = stats.ttest_rel(df["benchmark"], df[column])
    print(f"{column}: t-stat = {round(t_stat,4)}, p-value = {t_p_value:.4f}, {t_p_value>0.05}")
    sts_data['t-stat'].append(t_stat)
    sts_data['t-p-val'].append(t_p_value)
print("\nWilcoxon Results (vs. benchmark): p_val > 0.05 then same distribution")
for column in wiltest:
    t_stat, t_p_value = stats.wilcoxon(df["benchmark"], df[column])
    print(f"{column}: wil-stat = {round(t_stat,4)}, p-value = {t_p_value:.4f}, {t_p_value>0.05}")
sts_df = pd.DataFrame(sts_data)
```
